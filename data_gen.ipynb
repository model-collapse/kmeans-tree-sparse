```python
import numpy as np
import math
```

```python
pt1 = np.array([3, 10])
pt2 = np.array([10, 3])
```

```python
rng = np.random.default_rng()
def circle_sample(r):
    s = (rng.standard_normal(2) / 2 / math.pi) * r
    while (np.dot(s, s) > r * r):
        s = (rng.standard_normal(2) / 2 / math.pi) * r

    return s

```

```python
a = [(pt1 + circle_sample(2)).tolist() for i in range(10)] + [(pt2 + circle_sample(2)).tolist() for i in range(10)]
```

```python
a
```

```python
with open("bin/kmeans.jsonl", "w") as f:
    c = 0
    for l in a:
        print("%d\t{\"x\": %f, \"y\": %f}"%(c, l[0], l[1]), file=f)
        c += 1
```

```python
pt1 = np.array([3, 3])
pt2 = np.array([4, 4])
pt3 = np.array([3, 4])
pt4 = np.array([4, 3])
pt5 = np.array([10, 13])
pt6 = np.array([13, 10])

lst = [pt1, pt2, pt3, pt4, pt5, pt6]
pts = []
for pt in lst:
    pts += [(pt + circle_sample(1)).tolist() for i in range(10)]



```
